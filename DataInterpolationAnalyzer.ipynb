```python
import numpy as np
import matplotlib.pyplot as plt
import os

def plot_interpolation_and_derivative(file_path, num_points=500, figsize=(8, 6), dpi=300):
    """
    Read data from a file, interpolate it, compute its numerical derivative, and visualize both.

    Parameters:
        file_path (str): Path to the input data file (expected format: two columns for x and y).
        num_points (int): Number of points for interpolation.
        figsize (tuple): Figure size for plots (width, height).
        dpi (int): Resolution for saving plots.

    Returns:
        None: Generates and saves two plots (interpolated data and derivative).
    """
    # Check if the input file exists
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"Input file '{file_path}' not found.")

    # Load data from the file
    try:
        data = np.loadtxt(file_path)
        x = data[:, 0]
        y = data[:, 1]
    except Exception as e:
        raise ValueError(f"Error reading file '{file_path}': {str(e)}")

    # Perform linear interpolation
    x_interp = np.linspace(x[0], x[-1], num_points)
    y_interp = np.interp(x_interp, x, y)

    # Compute numerical derivative using np.gradient for better accuracy
    dy_dx = np.gradient(y_interp, x_interp)

    # Create plot for original and interpolated data
    plt.figure(figsize=figsize)
    plt.plot(x, y, 'ro', label='Original Data', markersize=5, alpha=0.7)
    plt.plot(x_interp, y_interp, 'b-', label='Interpolated Curve', linewidth=2)
    plt.xlabel('x', fontsize=12)
    plt.ylabel('y', fontsize=12)
    plt.title('Interpolated Function from Data', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend(fontsize=10)

    # Save the interpolation plot
    plt.savefig('interpolation_plot.png', dpi=dpi, bbox_inches='tight')
    plt.close()

    # Create plot for numerical derivative
    plt.figure(figsize=figsize)
    plt.plot(x_interp, dy_dx, 'g-', label='Numerical Derivative', linewidth=2)
    plt.xlabel('x', fontsize=12)
    plt.ylabel('dy/dx', fontsize=12)
    plt.title('Numerical Derivative of Interpolated Function', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend(fontsize=10)

    # Save the derivative plot
    plt.savefig('derivative_plot.png', dpi=dpi, bbox_inches='tight')
    plt.close()

    # Display both plots
    # Note: Displaying is optional; comment out if only saving is needed
    plt.figure(figsize=figsize)
    plt.plot(x, y, 'ro', label='Original Data', markersize=5, alpha=0.7)
    plt.plot(x_interp, y_interp, 'b-', label='Interpolated Curve', linewidth=2)
    plt.xlabel('x', fontsize=12)
    plt.ylabel('y', fontsize=12)
    plt.title('Interpolated Function from Data', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend(fontsize=10)
    plt.show()

    plt.figure(figsize=figsize)
    plt.plot(x_interp, dy_dx, 'g-', label='Numerical Derivative', linewidth=2)
    plt.xlabel('x', fontsize=12)
    plt.ylabel('dy/dx', fontsize=12)
    plt.title('Numerical Derivative of Interpolated Function', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend(fontsize=10)
    plt.show()

# Example usage
if __name__ == "__main__":
    file_path = "data-der.txt"
    try:
        plot_interpolation_and_derivative(file_path)
    except Exception as e:
        print(f"Error: {str(e)}")
```
